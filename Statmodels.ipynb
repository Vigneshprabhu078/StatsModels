```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
data=sm.datasets.get_rdataset("Guerry",package="HistData").data
```

```python
type(data)
```

```python
data.head()
```

```python
#fit ols regression use in
result=smf.ols("Lottery~Literacy+np.log(Pop1831)",data=data).fit()
```

```python
result.summary()
```

```python
nobs=100
X= np.random.random((nobs,2))
X= sm.add_constant(X)
```

```python
beta=np.array([1,0.1,10])
e=np.random.random(size=nobs)
y=X.dot(beta)+e
```

```python
result_2=sm.OLS(y,X).fit()
```

```python
result_2.summary()
```
